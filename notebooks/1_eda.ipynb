```python
# in the salary_range column, we have a lot of missing values, replace the double digits to thousands 15-20 to 15000-20000 and the single digits to thousands 1-2 to 15000-20000, also split the salary_range column into two columns salary_min and salary_max and leave the 0-0 as minimum and maximum salary 0 and 0
# Define a function to transform the salary range values
def clean_salary_range(value):
    if pd.isna(value) or value == "0-0":
        return "0-0"
    try:
        min_val, max_val = value.split("-")
        if len(min_val) == 2 and len(max_val) == 2:  # Double digits to thousands
            return f"{int(min_val)*1000}-{int(max_val)*1000}"
        elif len(min_val) == 1 and len(max_val) == 1:  # Single digits, assuming correction to "1000-2000"
            return f"{int(min_val)*1000}-{int(max_val)*2000}"
        else:
            return value  # Return as is if it doesn't fit the pattern
    except ValueError:  # Catch errors from invalid split or int conversion
        return "0-0"  # Treat unexpected format as "0-0"

# Apply the cleaning function
data['salary_range'] = data['salary_range'].apply(clean_salary_range)

# Split 'salary_range' into 'salary_min' and 'salary_max', then drop 'salary_range'
data[['salary_min', 'salary_max']] = data['salary_range'].str.split('-', expand=True)
data.drop('salary_range', axis=1, inplace=True)

# Convert 'salary_min' and 'salary_max' to integers, replacing any errors with 0
data['salary_min'] = pd.to_numeric(data['salary_min'], errors='coerce').fillna(0).astype(int)
data['salary_max'] = pd.to_numeric(data['salary_max'], errors='coerce').fillna(0).astype(int)
```

```python
# Split the 'location' column into 'country', 'state', and 'city'
data[['country', 'state', 'city']] = data['location'].str.split(',', expand=True, n=2)

# Fill missing values in the new location-based columns with "Unknown"
data[['country', 'state', 'city']] = data[['country', 'state', 'city']].fillna('Unknown')

# For the 'department' column, fill missing values with "Unknown"
data['department'] = data['department'].fillna('Unknown')

# Display the first few rows to verify the changes
data[['location', 'country', 'state', 'city', 'department']].head()
```
